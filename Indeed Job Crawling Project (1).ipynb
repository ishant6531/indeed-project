# Indeed Job Crawling project

```python
# Loading  Necessary  library 

from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
import warnings
warnings.filterwarnings('ignore')
import time
import string

driver=webdriver.Chrome(r"C:\Users\thunderbolt\sof\chromedriver.exe")

Job_Title = []
Company = []
Rating = []
Location = []
Short_Description = []
Days_Posted = []
Salary = []
Brief_Description = []
```

```python
# Getting   all the job  details


for i in range(0,50,10):

    driver.get("https://in.indeed.com/jobs?q=Data+Scientist+&l=Gurgaon%2C+Haryana&start="+ str(i))

    time.sleep(5)

    all_jobs = driver.find_elements_by_class_name('result')

    for job in all_jobs:
    
        result_html= job.get_attribute('innerHTML')
        soup = BeautifulSoup(result_html,'html.parser')
    
    
        try:
            title = soup.find("a" ,class_="jobtitle").text.replace('\n',' ')
            Job_Title.append(title)
        except:
            Job_Title.append('---')
        
    
    
        try:
            company = soup.find("span" ,class_="company").text.replace('\n',' ')
            Company.append(company)
        except:
            Company.append('---')
        

        try:
            rating = soup.find("span" ,class_="ratingsContent").text.replace('\n',' ')
            Rating.append(rating)
        except:
            Rating.append('---')
        
        try:
            location = soup.find("span" ,class_="location accessible-contrast-color-location").text.replace('\n',' ')
            Location.append(location)
        except:
            Location.append('---')
        
        
        try:
            salary = soup.find("span" ,class_="salaryText").text.replace('\n',' ')
            Salary.append(salary)
        except:
            Salary.append('Not Mentioned')
        
        
        try:
            short = soup.find("div" ,class_="summary").text.replace('\n',' ')
            Short_Description.append(short)
        except:
            Short_Description.append('---')
        
        
        try:
            days_posted = soup.find("span" ,class_="date").text.replace('\n',' ')
            Days_Posted.append(days_posted)
        except:
            Days_Posted.append('---')
        
       
        
        sum_div = job.find_elements_by_class_name("summary")[0]
        sum_div.click()
        
        time.sleep(5)
        
        
        try:
            job_desc = driver.find_element_by_id("vjs-desc").text.replace('\n',' ')
            Brief_Description.append(job_desc)
        except:
            Brief_Description.append('----')
        
       
```

```python
print(len(Job_Title) , len(Company) ,len(Rating) ,len(Location) ,len(Days_Posted) ,len(Salary),len(Short_Description),len(Days_Posted ),len(Brief_Description))
```

```python
# Getting Final Data of Data Scientist Job  for the Gurugram location.

Data_Science =pd.DataFrame({'Job Title': Job_Title , 'Company' : Company , 'Rating' : Rating ,'Location' :Location , 'Days Posted' : Days_Posted , 'Salary' : Salary , 'Short Description' :Short_Description , 'Brief Description' : Brief_Description})
Data_Science 
```
